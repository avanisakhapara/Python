```python
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(7)
print(x)

print(plt.style.available)
#plt.style.use('fivethirtyeight')

# plt.plot(x,(x*x),color='blue',marker='o',label='square of values')#'bo--'
plt.plot(x,(x*x),'bo--',label="square of x")
plt.plot(x,x,color='red',marker='v',linestyle='dashed',label='linear (y=x)')
plt.plot(x,(x*x*x),color='green',linestyle='--',label='cube of x')

# y_point=x*x

a_point = 50


# filling plots
#plt.fill_between(x,(x*x*x),a_point,alpha=0.25)

#plt.fill_between(x,(x*x*x),a_point,where=((x*x*x) > a_point),alpha=0.25, interpolate = True, color='red',label="greater than 50")

#plt.fill_between(x,(x*x*x),a_point,where=((x*x*x) <= a_point),alpha=0.25, interpolate = True,color='green',label="less than or equal to 50")
plt.fill_between(x,(x*x*x),(x*x),alpha=0.10)

plt.xlabel('number')
plt.ylabel('square')
plt.title('Example of plotting lines')

plt.legend()
plt.savefig('square_fig.jpg')
plt.show()
```

```python
#pie charts

import matplotlib.pyplot as plt

# plt.style.use('seaborn')

slices = [120,30,50,80,90]
labels = ['python','javascript','C#','php','java']
colors = [ '#ff8883','#bc345d','#ef238c','#783db2','#45b34d']
explode=[0,0.5,0,0,0]

plt.pie(slices, labels=labels,colors=colors,wedgeprops={'edgecolor':'blue'},startangle=230,autopct='%1.1f%%',explode=explode)
#wedgeprops={'edgecolor':'black'},startangle=230,autopct='%1.1f%%',shadow=True,explode=explode

plt.legend(loc='best')
plt.title("Pie chart example")
#plt.tight_layout()
plt.show()
```

```python
#stack plots /area plots
import matplotlib.pyplot as plt

#plt.style.use('fivethirtyeight')

days = [1,2,3,4,5,6]

dev1 = [2,2,4,4,5,8]
dev2 = [1,1,1,1,2,2]
dev3 = [2,2,3,3,4,5]

labels =['developer1','developer2','developer3']

plt.stackplot(days,dev1,dev2,dev3,labels=labels)

plt.legend(loc='upper left')

plt.title("Stack Plot example")
plt.tight_layout()
plt.show()
```

```python
#histogram

import matplotlib.pyplot as plt

#plt.style.use('fivethirtyeight')


marks = [35,20,47,69,50,78,69,61,50]

#plt.hist(marks,bins=5,edgecolor='black')

bins=[20,30,40,50,60,70]
plt.hist(marks,bins=bins,edgecolor='black')

#plt.legend()

plt.title("Histogram Example")
plt.xlabel('marks')
plt.ylabel('no. of students')
plt.tight_layout()
plt.show()
```

```python
#scatter plot - multivariate outliers

import matplotlib.pyplot as plt

#plt.style.use('fivethirtyeight')


ESE_marks = [35,20,47,69,50,78,69,61,50]
CA_marks =  [25,34,38,44,31,47,43,40,40]

plt.scatter(ESE_marks,CA_marks)
plt.scatter(ESE_marks,CA_marks,s=100,c='green',edgecolor='black',linewidth=3,alpha=0.40)



#plt.legend()

plt.title("Scatter Plot Example")
plt.xlabel('ESE_marks')
plt.ylabel('CA_marks')
plt.tight_layout()
plt.show()
```

```python
#box plot #univariate outliers
#min, max, median, ist quantile, 3rd quantile
import matplotlib.pyplot as plt

ages = [10,30,45,23,50,34,88,15,18]#[10,15,18,23,[30],34,45,50,88] #median=30,1st =18 , 3rd =45
plt.boxplot(ages,patch_artist=True,vert=0)
#vert=0,patch_artist=True
plt.show()
```

```python
#Object Oriented Approach
import matplotlib.pyplot as plt
import numpy as np

x=np.array([2,4,5,6,7,9])

fig=plt.figure() #returns figure object
ax=fig.add_axes([0,0,1,1]) #return Axes object
l1=ax.plot(x,(x*x))
l2=ax.plot(x,x,'rv--')
ax.set_xlim(0,12)
ax.set_ylim(0,120)
ax.spines['right'].set_color(None)
#ax.spines['top'].set_color(None)
ax.spines['bottom'].set_color('green')
ax.spines['bottom'].set_linewidth(3)
ax.set_xlabel('number')
ax.set_ylabel('Square of number')
ax.set_title('Number Vs Square of Number')
ax.legend(labels=['linear','exponential'],loc='best')
plt.show()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)#having multiple subplots within a single figure ->returns figure and axes
x = np.arange(10)

ax[0, 0].plot(x,x, 'g') #row=0, col=0
ax[0, 0].grid(True)

ax[1, 0].plot(x,(x*x), 'k') #row=1, col=0

ax[0, 1].plot(x,(x*x*x), 'b') #row=0, col=1
ax[0, 1].grid(True,color='g',linestyle='--',linewidth=3)


ax[1, 1].plot(x, (x*5), 'r') #row=1, col=1
fig.show()
```
